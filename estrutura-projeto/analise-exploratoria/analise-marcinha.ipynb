Visualizações dos tipos de obesidades que são 3 tipos de obesidades:

A obesidade é comumente classificada de acordo com o IMC, que é uma medida que relaciona o peso e a altura de uma pessoa. As categorias padrão de classificação de acordo com o IMC são:

Abaixo do peso: IMC menor que 18,5
Peso normal: IMC entre 18,5 e 24,9
Sobrepeso: IMC entre 25 e 29,9
Obesidade: IMC igual ou superior a 30
Obesidade Tipo I (ou obesidade leve): IMC entre 30 e 34,9
Obesidade Tipo II (ou obesidade moderada): IMC entre 35 e 39,9
Obesidade Tipo III (ou obesidade mórbida): IMC igual ou superior a 40.

```python




# Contagem dos tipos de obesidade na coluna 'NObeyesdad'
tipos = df_tipos_obesidade['NObeyesdad'].value_counts()

# Criar um gráfico de barras simples
tipos.plot(kind='bar', color='skyblue')
plt.xlabel('Tipos de Obesidade')
plt.ylabel('Contagem')
plt.title('Distribuição dos Tipos de Obesidade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Filtrar as linhas com IMC maior que 30
bmi_maior_que_30 = df_classificacao_obesidade[df_classificacao_obesidade['BMI'] > 30]

# Criar um histograma dos valores de IMC acima de 30
plt.hist(bmi_maior_que_30['BMI'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Índice de Massa Corporal (IMC)')
plt.ylabel('Frequência')
plt.title('Distribuição de IMC acima de 30')
plt.grid(True)
plt.show()
```
