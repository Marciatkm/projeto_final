<h1>Visualizações dos tipos de obesidades ;<h1>



```python
#QUAL O  TIPO DE OBESIDADE MAIS FREQUENTE?

tipos = df_tipos_obesidade['NObeyesdad'].value_counts()
tipos

#O TIPO 1 É O MAIS FREQUENTE.
```

```python
# Contagem de pessoas que prática atividade física
atividade_fisica = df_tipos_obesidade['SCC'].value_counts()
atividade_fisica
```

```python
# Contagem de pessoas que prática atividade física
atividade_fisica = df_tipos_obesidade['SCC'].value_counts()

# Criando um gráfico de barras simples
plt.bar(atividade_fisica.index, atividade_fisica.values)
plt.ylabel('Contagem')
plt.title('   Prática de Atividade Física')
plt.show()
```

```python
#QUAL GENERO ESTÁ MAIS OBESOS?
genero = df_tipos_obesidade['Gender'].value_counts()
genero
# O GÊNERO MASCULINO É O MAIS OBESO.
```

```python
#QUAL A IDADE DOS OBESOS?
Idades = df_tipos_obesidade['Age'].value_counts()
Idades
#A IDADE É DE 18 ANOS 
```

```python
# Contagem dos tipos de obesidade na coluna 'NObeyesdad'
tipos = df_tipos_obesidade['NObeyesdad'].value_counts()

# Criar um gráfico de barras simples
tipos.plot(kind='bar', color='RED')
plt.xlabel('Tipos de Obesidade')
plt.ylabel('Contagem')
plt.title('Distribuição dos Tipos de Obesidade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h1> Visualizações  da classificaÇao obesidades;<h1>

```python
# OS NÚMEROS DOS ÍNDICES DE MASSA CORPORAL (IMC)
imc = df_classificacao_obesidade['BMI'].value_counts()
print(imc)
```

```python
# Filtrar as linhas com índice de massa corporal (IMC) maior que 30;
bmi_maior_que_30 = df_classificacao_obesidade[df_classificacao_obesidade['BMI'] > 30]

# Mostrar o DataFrame resultante
print(bmi_maior_que_30)

```

```python

# Filtrar as linhas com IMC maior que 30
bmi_maior_que_30 = df_classificacao_obesidade[df_classificacao_obesidade['BMI'] > 30]

# Criar um histograma dos valores de IMC acima de 30
plt.hist(bmi_maior_que_30['BMI'], bins=20, color='skyblue', edgecolor='black')

plt.title('Índice de Massa Corporal (IMC) acima de 30')
plt.grid(True)
plt.show()
```
